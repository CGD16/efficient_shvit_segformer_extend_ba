```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from matplotlib.ticker import MultipleLocator, AutoMinorLocator, LogLocator, NullFormatter
```

```python
def plot_acc_vs_memory(filename, x_MVT, mean_values_MVT, std_devs_MVT,
            x_SHViT_B0, mean_values_SHViT_B0, std_devs_SHViT_B0,
            x_SHViT_B5, mean_values_SHViT_B5, std_devs_SHViT_B5,
            x_SHViT_SW_B0, mean_values_SHViT_SW_B0, std_devs_SHViT_SW_B0,
            x_SHViT_SW_B5, mean_values_SHViT_SW_B5, std_devs_SHViT_SW_B5,
            capsize=5, markersize=8):
    
    # Colors for individual points
    colors_MVT = ['lightcoral', 'salmon', 'crimson', 'firebrick', 'darkred', 'indianred']
    labels_MVT = ['B0', 'B1', 'B2', 'B3', 'B4', 'B5']
    colors_SHViT_B0 = ['lightskyblue', 'cornflowerblue', 'royalblue', 'mediumblue', 'midnightblue']
    labels_SHViT_B0 = ['B0/S1', 'B0/S2', 'B0/S3', 'B0/S4']
    colors_SHViT_B5 = ['lightgreen', 'limegreen', 'seagreen', 'forestgreen', 'darkgreen']
    labels_SHViT_B5 = ['B5/S1', 'B5/S2', 'B5/S3', 'B5/S4']
    colors_SHViT_SW_B0 = ['peachpuff', 'darksalmon', 'coral', 'orangered', 'chocolate']
    labels_SHViT_SW_B0 = ['B0/S1', 'B0/S2', 'B0/S3', 'B0/S4']
    colors_SHViT_SW_B5 = ['lavenderblush', 'plum', 'mediumorchid', 'darkorchid', 'indigo']
    labels_SHViT_SW_B5 = ['B5/S1', 'B5/S2', 'B5/S3', 'B5/S4']
    
    # Creating the figure with specific size
    fig, ax = plt.subplots(figsize=(13, 6))
    
    
    # Plotting with error bars for the first dataset
    for xi, yi, si, color, lbl in zip(x_MVT, mean_values_MVT, std_devs_MVT, colors_MVT, labels_MVT):
        ax.errorbar(
            xi, yi, yerr=si, fmt='o', capsize=capsize,
            ecolor='black', elinewidth=0.5,
            markerfacecolor=color, markeredgecolor=color, markersize=markersize,
            label=f"{lbl} (MVT)"
        )
    
    for xi, yi, si, color, lbl in zip(x_SHViT_B0, mean_values_SHViT_B0, std_devs_SHViT_B0, colors_SHViT_B0, labels_SHViT_B0):
        ax.errorbar(
            xi, yi, yerr=si, fmt='o', capsize=capsize,
            ecolor='black', elinewidth=0.5,
            markerfacecolor=color, markeredgecolor=color, markersize=markersize,
            label=f"{lbl} (SHViT)"
        )
    
    for xi, yi, si, color, lbl in zip(x_SHViT_B5, mean_values_SHViT_B5, std_devs_SHViT_B5, colors_SHViT_B5, labels_SHViT_B5):
        ax.errorbar(
            xi, yi, yerr=si, fmt='o', capsize=capsize,
            ecolor='black', elinewidth=0.5,
            markerfacecolor=color, markeredgecolor=color, markersize=markersize,
            label=f"{lbl} (SHViT)"
        )
    
    for xi, yi, si, color, lbl in zip(x_SHViT_SW_B0, mean_values_SHViT_SW_B0, std_devs_SHViT_SW_B0, colors_SHViT_SW_B0, labels_SHViT_SW_B0):
        ax.errorbar(
            xi, yi, yerr=si, fmt='o', capsize=capsize,
            ecolor='black', elinewidth=0.5,
            markerfacecolor=color, markeredgecolor=color, markersize=markersize,
            label=f"{lbl} (SHViT SW)"
        )
    
    for xi, yi, si, color, lbl in zip(x_SHViT_SW_B5, mean_values_SHViT_SW_B5, std_devs_SHViT_SW_B5, colors_SHViT_SW_B5, labels_SHViT_SW_B5):
        ax.errorbar(
            xi, yi, yerr=si, fmt='o', capsize=capsize,
            ecolor='black', elinewidth=0.5,
            markerfacecolor=color, markeredgecolor=color, markersize=markersize,
            label=f"{lbl} (SHViT SW)"
        )
    
    # Customizing the plot
    # ax.set_title('3D Segformer with different backbones and parameter sets', size=18)
    ax.set_xlabel('Memory Consumption (in GB)', size=16)
    ax.set_ylabel('F1-score (fraction)', size=16)

    ax.set_xlim(0, 25)
    ax.set_ylim(0.1, 0.8)
    ax.yaxis.set_major_locator(MultipleLocator(0.1))
    ax.xaxis.set_minor_locator(AutoMinorLocator())
    ax.yaxis.set_minor_locator(AutoMinorLocator())
    ax.tick_params(axis='both', which='minor', labelbottom=False, labelleft=False)
    ax.tick_params(axis='both', which='minor', length=3)

    ax.grid(True, linestyle='dotted')  # Set grid line style to dotted
    ax.tick_params(axis='both', labelsize=14)  # set fontsize to 14
    
    # Adding a legend
    # Creating handles for custom legend layout
    legend_elements = [
        # — 6 MVT —
        Line2D([0],[0],marker='o',color='w',label='B0 (MVT)', markerfacecolor='lightcoral', markersize=markersize),
        Line2D([0],[0],marker='o',color='w',label='B1 (MVT)', markerfacecolor='salmon',    markersize=markersize),
        Line2D([0],[0],marker='o',color='w',label='B2 (MVT)', markerfacecolor='crimson',   markersize=markersize),
        Line2D([0],[0],marker='o',color='w',label='B3 (MVT)', markerfacecolor='firebrick', markersize=markersize),
        Line2D([0],[0],marker='o',color='w',label='B4 (MVT)', markerfacecolor='darkred',   markersize=markersize),
        Line2D([0],[0],marker='o',color='w',label='B5 (MVT)', markerfacecolor='indianred', markersize=markersize),
        # — 16 SHViT & SHViT SW —
        Line2D([0],[0],marker='o',color='w',label='B0/S1 (SHViT)',    markerfacecolor='lightskyblue', markersize=markersize),
        Line2D([0],[0],marker='o',color='w',label='B0/S2 (SHViT)',    markerfacecolor='cornflowerblue', markersize=markersize),
        Line2D([0],[0],marker='o',color='w',label='B0/S3 (SHViT)',    markerfacecolor='royalblue',      markersize=markersize),
        Line2D([0],[0],marker='o',color='w',label='B0/S4 (SHViT)',    markerfacecolor='mediumblue',     markersize=markersize),
        Line2D([0],[0],marker='o',color='w',label='B5/S1 (SHViT)',    markerfacecolor='lightgreen',     markersize=markersize),
        Line2D([0],[0],marker='o',color='w',label='B5/S2 (SHViT)',    markerfacecolor='limegreen',      markersize=markersize),
        Line2D([0],[0],marker='o',color='w',label='B5/S3 (SHViT)',    markerfacecolor='seagreen',       markersize=markersize),
        Line2D([0],[0],marker='o',color='w',label='B5/S4 (SHViT)',    markerfacecolor='forestgreen',    markersize=markersize),
        Line2D([0],[0],marker='o',color='w',label='B0/S1 (SHViT CSA)', markerfacecolor='peachpuff',     markersize=markersize),
        Line2D([0],[0],marker='o',color='w',label='B0/S2 (SHViT CSA)', markerfacecolor='darksalmon',    markersize=markersize),
        Line2D([0],[0],marker='o',color='w',label='B0/S3 (SHViT CSA)', markerfacecolor='coral',         markersize=markersize),
        Line2D([0],[0],marker='o',color='w',label='B0/S4 (SHViT CSA)', markerfacecolor='orangered',     markersize=markersize),
        Line2D([0],[0],marker='o',color='w',label='B5/S1 (SHViT CSA)', markerfacecolor='lavenderblush', markersize=markersize),
        Line2D([0],[0],marker='o',color='w',label='B5/S2 (SHViT CSA)', markerfacecolor='plum',          markersize=markersize),
        Line2D([0],[0],marker='o',color='w',label='B5/S3 (SHViT CSA)', markerfacecolor='mediumorchid',  markersize=markersize),
        Line2D([0],[0],marker='o',color='w',label='B5/S4 (SHViT CSA)', markerfacecolor='darkorchid',    markersize=markersize),
    ]

    # 2) create a completely invisible “dummy” handle:
    dummy = Line2D([0],[0],
                marker='o',
                linestyle='',
                markerfacecolor='none',
                markeredgecolor='none',
                label='')      # empty label

    # 3) insert exactly two dummies at the break-points:
    #    – one just before the SHViT block to pad column-1 down to 6  
    #    – one just before the last SHViT-SW so that col-2 also has blanks
    legend_elements.insert(10, dummy)
    legend_elements.insert(11, dummy)
    legend_elements.insert(16, dummy)
    legend_elements.insert(17, dummy)
    legend_elements.insert(22, dummy)
    legend_elements.insert(23, dummy)
    legend_elements.insert(28, dummy)

    # 4) draw a single legend with 4 columns:
    ax.legend(handles=legend_elements, ncol=5, loc='best', fontsize='large', columnspacing=0.5)
    plt.savefig(filename, dpi=300, bbox_inches='tight', pad_inches=0)
    
    # Show the plot
    plt.show()
```

```python
colors = [
    "#DC143C",  # Crimson
    "#FF8C00",  # Dark Orange
    "#FFD700",  # Gold
    "#32CD32",  # Lime Green
    "#00BFFF",  # Deep Sky Blue
    "#C71585",  # Medium Violet Red
    "#8B4513",   # Saddle Brown
    "#000000",  # Black
]

def plot_acc_vs_vol_memory(filename, x, mean_values, std_devs,
            capsize=5, markersize=8):
    
    # Colors for individual points
    # colors = ['lightskyblue', 'cornflowerblue', 'royalblue', 'blue', 'mediumblue', 'midnightblue']
    labels = ['96x96x96', '128x128x128', '160x160x160', '192x192x192', '224x224x224', '256x256x256', '288x288x288', '320x320x320']
    
    # Creating the figure with specific size
    fig, ax = plt.subplots(figsize=(13, 6))
    
    
    # Plotting with error bars for the first dataset
    for xi, yi, si, color, lbl in zip(x, mean_values, std_devs, colors, labels):
        ax.errorbar(
            xi, yi, yerr=si, fmt='o', capsize=capsize,
            ecolor='black', elinewidth=0.5,
            markerfacecolor=color, markeredgecolor=color, markersize=markersize,
            label=f"{lbl}"
        )
    
    # Customizing the plot
    # ax.set_title('3D Segformer with different backbones and parameter sets', size=18)
    ax.set_xlabel('Memory Consumption (in GB)', size=16)
    ax.set_ylabel('F1-score (fraction)', size=16)

    ax.set_xlim(0, 25)
    ax.set_ylim(0.3, 0.6)
    ax.yaxis.set_major_locator(MultipleLocator(0.1))
    ax.xaxis.set_minor_locator(AutoMinorLocator())
    ax.yaxis.set_minor_locator(AutoMinorLocator())
    ax.tick_params(axis='both', which='minor', labelbottom=False, labelleft=False)
    ax.tick_params(axis='both', which='minor', length=3)

    ax.grid(True, linestyle='dotted')  # Set grid line style to dotted
    ax.tick_params(axis='both', labelsize=14)  # set fontsize to 14
    
    # Adding a legend
    # Creating handles for custom legend layout
    legend_elements = [
        Line2D([0],[0],marker='o',color='w',label=labels[0], markerfacecolor=colors[0], markersize=markersize),
        Line2D([0],[0],marker='o',color='w',label=labels[1], markerfacecolor=colors[1], markersize=markersize),
        Line2D([0],[0],marker='o',color='w',label=labels[2], markerfacecolor=colors[2], markersize=markersize),
        Line2D([0],[0],marker='o',color='w',label=labels[3], markerfacecolor=colors[3], markersize=markersize),
        Line2D([0],[0],marker='o',color='w',label=labels[4], markerfacecolor=colors[4], markersize=markersize),
        Line2D([0],[0],marker='o',color='w',label=labels[5], markerfacecolor=colors[5], markersize=markersize),
        Line2D([0],[0],marker='o',color='w',label=labels[6], markerfacecolor=colors[6], markersize=markersize),
        Line2D([0],[0],marker='o',color='w',label=labels[7], markerfacecolor=colors[7], markersize=markersize),
    ]

    ax.legend(handles=legend_elements, loc='best', fontsize='large', columnspacing=0.5)
    plt.savefig(filename, dpi=300, bbox_inches='tight', pad_inches=0)
        
    # Show the plot
    plt.show()
```
